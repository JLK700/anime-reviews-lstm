```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
reviews = pd.read_json('../MyAnimeList Scraper/anime_reviews.json')
```

```python
reviews = reviews.transpose()
```

```python
reviews.score.map(int).mean()
```

```python
scores = reviews.groupby('score').count()
scores.index = scores.index.map(int)
plt.figure(figsize=(10, 7))
plt.bar(scores.index[1:], scores.values.flatten()[1:])
plt.xticks(scores.index[1:])
plt.title('Reviews', size=28, pad=20)
plt.xlabel('score', fontsize=18, labelpad=20)
plt.ylabel('amount', fontsize=18, labelpad=20)
plt.show()
```

```python
print(f'mean: {scores.text.mean()}')
print(f'std: {scores.text.std()}')
```

```python
reviews['length'] = reviews['text'].str.len().to_frame(name='length')
reviews
```

```python
len_dict = {'very_short (< 500)': reviews.length[(reviews.length < 500)].count(),
            'short (< 1000)': reviews.length[(reviews.length >= 500) & (reviews.length < 1000)].count(),
            'medium_short (< 2000)': reviews.length[(reviews.length >= 1000) & (reviews.length < 2000)].count(),
            'medium (< 3000)': reviews.length[(reviews.length >= 2000) & (reviews.length < 3000)].count(),
            'medium_long (< 5000)': reviews.length[(reviews.length >= 3000) & (reviews.length < 5000)].count(),
            'long (< 10000)': reviews.length[(reviews.length >= 5000) & (reviews.length < 10000)].count(),
            'very_long (> 10000)': reviews.length[(reviews.length >= 10000)].count()}
```

```python
plt.figure(figsize=(16, 5))
plt.bar(len_dict.keys(), len_dict.values())
plt.title('Reviews lengths', fontsize=28, pad=20)
plt.xlabel('length (in characters amount)', fontsize=18, labelpad=20)
plt.ylabel('amount', fontsize=18, labelpad=20)

plt.show()
```

```python
plt.figure(figsize=(16, 5))
plt.bar(len_score_dict.keys(), len_score_dict.values())
plt.title('Reviews mean scores by length', fontsize=28, pad=20)
plt.xlabel('length (in characters amount)', fontsize=18, labelpad=20)
plt.ylabel('mean score', fontsize=18, labelpad=20)
plt.ylim(6,7.3)
plt.yticks(np.arange(6, 7.3, step=0.1))

plt.show()
```
